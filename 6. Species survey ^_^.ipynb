## Joining 2 dataframes

- combining dataframes using a common column is called joining
- so here species_id is the common col
- hence species will be taken as the lookup table 

```python
survey_sub
```

```python
species_sub= species.head(20)
species_sub
#sample lookup table
```

```python
len(species)
#all species_id's in species are unique 
# there are 54 unique species_id for the species csv file 
```

```python
#merged inner
merge_in = pd.merge(left=survey_sub,right=species_sub, left_on='species_id', right_on='species_id')
#left and right have the same column name
merge_in
#now the rows with the same species id as the species_sub will get joined hence nl and dm
# genus, species and taxa has been joined too
```

```python
#merge_left join
#to join 2 dataframe without losing any information from the left dataframe 
merge_left= pd.merge(left= survey_sub, right=species_sub, how='left', left_on='species_id', right_on='species_id')
merge_left
#like so
#the rows without matching species_id gets filled with NaN values
```

```python
merge_left[merge_left.genus.isnull()]
#the missing values
```

```python
#merge_right
#to join 2 dataframe such that the unmatching ones in the right df are kept the same 
#and the ones in left are discarded
merge_right= pd.merge(left= survey_sub, right=species_sub, how='right', left_on='species_id', right_on='species_id')
merge_right
#likeso
```

## list of all species and their corresponding taxa,  with hindfootlength >= 55
